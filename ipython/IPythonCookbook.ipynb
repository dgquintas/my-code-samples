# Getting started with exploratory data analysis in IPython (p. 22)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "http://donnees.ville.montreal.qc.ca/storage/f/2014-01-20T20%3A48%3A50.296Z/2013.csv"
df = pd.read_csv(url, index_col='Date', parse_dates=True, dayfirst=True)
df.head(2)
```

```python
df.describe()
```

```python
df[['Berri1', 'PierDup']].plot()
```

```python
days = np.array(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
df['Weekday'] = days[df.index.weekday]
```

```python
df_week = df.groupby('Weekday').sum()
df_week
```

```python
df_week.ix[days].plot(lw=3)
plt.ylim(0);
```

```python
from IPython.html.widgets import interact

@interact
def plot(n=(1,30)):
    pd.rolling_mean(df['Berri1'], n).dropna().plot()
    plt.ylim(0, 8000)
    plt.show()
```

# Introducing the multidimensional array in NumPy for fast array computations

```python
import random
import numpy as np

# Show only three decimals in the Python output
%precision 3
```

Two lists with 1 million elements

```python
n = 1000000
x = [random.random() for _ in range(n)]
y = [random.random() for _ in range(n)]
```

```python
%timeit [x[i] + y[i] for i in range(n)]
```

Now with NumPy

```python
xa = np.array(x)
ya = np.array(y)
%timeit xa + ya
```

Damn that was fast(er)!

```python
%timeit sum(x)  # Pure Python
%timeit np.sum(xa)  # NumPy
```

```python
%timeit [abs(x[i] - y[j]) for i in range(1000) for j in range(1000)]  # We use only 1000 to avoid waiting
%timeit np.abs(xa[:1000, None] - ya[:1000])
```

# Machine Learning

## Getting started with scikit-learn

```python
import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# A deterministic non-linear function underlying our generative model
f = lambda x: np.exp(3 * x)

x_tr = np.linspace(0., 2, 200)
y_tr = f(x_tr)

x = np.array([0, .1, .2, .5, .8, .9, 1])
y = f(x) + np.random.randn(len(x))
```

```python
plt.plot(x_tr[:100], y_tr[:100], '--k')
plt.plot(x, y, 'ok', ms=10)
```

```python
# Create the linear model
lr = lm.LinearRegression()

# Train the model on our training dataset
lr.fit(x[:, np.newaxis], y)

# Predict points with our trained model
y_lr = lr.predict(x_tr[:, np.newaxis])
```

```python
plt.plot(x_tr, y_tr, '--k')
plt.plot(x_tr, y_lr, 'g')
plt.plot(x, y, 'ok', ms=10)
plt.xlim(0, 1)
plt.ylim(y.min()-1, y.max()+1)
plt.title("Linear regression")
```
